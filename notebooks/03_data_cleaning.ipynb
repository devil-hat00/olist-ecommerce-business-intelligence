# Olist E-Commerce Business Intelligence

## Phase 3 — Data Cleaning

### Objective

The objective of this notebook is to clean and prepare the Olist Brazilian E-Commerce dataset for exploratory data analysis and predictive modeling.

Cleaning operations include:

- Handling missing values
- Removing unnecessary duplicates
- Converting data types
- Standardizing data
- Validating cleaned datasets
- Saving processed datasets

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
PROJECT_ROOT = Path().resolve().parent

RAW_PATH = PROJECT_ROOT / "data" / "raw"

PROCESSED_PATH = PROJECT_ROOT / "data" / "processed"
```

```python
customers = pd.read_csv(RAW_PATH / "olist_customers_dataset.csv")

geolocation = pd.read_csv(RAW_PATH / "olist_geolocation_dataset.csv")

order_items = pd.read_csv(RAW_PATH / "olist_order_items_dataset.csv")

payments = pd.read_csv(RAW_PATH / "olist_order_payments_dataset.csv")

reviews = pd.read_csv(RAW_PATH / "olist_order_reviews_dataset.csv")

orders = pd.read_csv(RAW_PATH / "olist_orders_dataset.csv")

products = pd.read_csv(RAW_PATH / "olist_products_dataset.csv")

sellers = pd.read_csv(RAW_PATH / "olist_sellers_dataset.csv")

translation = pd.read_csv(RAW_PATH / "product_category_name_translation.csv")
```

```python
customers_clean = customers.copy()

geolocation_clean = geolocation.copy()

order_items_clean = order_items.copy()

payments_clean = payments.copy()

reviews_clean = reviews.copy()

orders_clean = orders.copy()

products_clean = products.copy()

sellers_clean = sellers.copy()

translation_clean = translation.copy()
```

## Why Create Copies?

The raw datasets remain unchanged throughout the project.

All cleaning operations are performed on independent copies to ensure reproducibility and preserve the original source data.

```python
products_clean["product_category_name"] = (
    products_clean["product_category_name"]
    .fillna("Unknown")
)
```

```python
numeric_columns = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for column in numeric_columns:
    products_clean[column] = (
        products_clean[column]
        .fillna(products_clean[column].median())
    )
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for column in date_columns:
    orders_clean[column] = pd.to_datetime(
        orders_clean[column],
        errors="coerce"
    )
```

```python
for name, df in {
    "Products": products_clean,
    "Orders": orders_clean,
    "Reviews": reviews_clean
}.items():

    print("=" * 70)
    print(name)
    print(df.isnull().sum())
```

```python
customers_clean.to_csv(
    PROCESSED_PATH / "customers_clean.csv",
    index=False
)

geolocation_clean.to_csv(
    PROCESSED_PATH / "geolocation_clean.csv",
    index=False
)

order_items_clean.to_csv(
    PROCESSED_PATH / "order_items_clean.csv",
    index=False
)

payments_clean.to_csv(
    PROCESSED_PATH / "payments_clean.csv",
    index=False
)

reviews_clean.to_csv(
    PROCESSED_PATH / "reviews_clean.csv",
    index=False
)

orders_clean.to_csv(
    PROCESSED_PATH / "orders_clean.csv",
    index=False
)

products_clean.to_csv(
    PROCESSED_PATH / "products_clean.csv",
    index=False
)

sellers_clean.to_csv(
    PROCESSED_PATH / "sellers_clean.csv",
    index=False
)

translation_clean.to_csv(
    PROCESSED_PATH / "translation_clean.csv",
    index=False
)
```

```python
cleaned_datasets = {
    "Customers": customers_clean,
    "Geolocation": geolocation_clean,
    "Order Items": order_items_clean,
    "Payments": payments_clean,
    "Reviews": reviews_clean,
    "Orders": orders_clean,
    "Products": products_clean,
    "Sellers": sellers_clean,
    "Translation": translation_clean
}

for name, df in cleaned_datasets.items():
    print("=" * 70)
    print(name)
    print(f"Shape: {df.shape}")
    print(f"Missing Values: {df.isnull().sum().sum()}")
```

# Cleaning Validation Summary

The cleaning process successfully addressed issues that required correction while preserving missing values that carry business meaning.

## Successfully Cleaned

- Product metadata missing values were handled.
- Product dimensions were imputed using the median.
- Timestamp columns were converted to datetime.
- All datasets were successfully exported to the processed directory.

## Intentionally Preserved Missing Values

### Reviews

Missing review titles and messages were retained because customers are not required to provide written feedback. These missing values represent legitimate user behavior.

### Orders

Missing delivery-related timestamps were retained because they may correspond to cancelled, pending, or undelivered orders. Preserving these values maintains the integrity of the business process.

## Final Assessment

The cleaned datasets are analysis-ready while maintaining the original business semantics and relational integrity.
