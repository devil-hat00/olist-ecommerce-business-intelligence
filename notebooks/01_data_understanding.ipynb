## Phase 1 - Data Understanding

### Objective
Understand the structure, relationships and quality of the Olist Brazilian E-Commerce dataset before performing cleaning or analysis

This notebook covers:
- Loading datasets
- Understanding schema
- Identifying keys
- Initial profiling
- Relationship mapping

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
PROJECT_ROOT = Path().resolve().parent

DATA_PATH = PROJECT_ROOT / "data" / "raw"
```

```python
customers = pd.read_csv(DATA_PATH / "olist_customers_dataset.csv")

geolocation = pd.read_csv(DATA_PATH / "olist_geolocation_dataset.csv")

order_items = pd.read_csv(DATA_PATH / "olist_order_items_dataset.csv")

payments = pd.read_csv(DATA_PATH / "olist_order_payments_dataset.csv")

reviews = pd.read_csv(DATA_PATH / "olist_order_reviews_dataset.csv")

orders = pd.read_csv(DATA_PATH / "olist_orders_dataset.csv")

products = pd.read_csv(DATA_PATH / "olist_products_dataset.csv")

sellers = pd.read_csv(DATA_PATH / "olist_sellers_dataset.csv")

translation = pd.read_csv(DATA_PATH / "product_category_name_translation.csv")
```

```python
datasets = {
    "Customers": customers,
    "Geolocation": geolocation,
    "Order Items": order_items,
    "Payments": payments,
    "Reviews": reviews,
    "Orders": orders,
    "Products": products,
    "Sellers": sellers,
    "Translation": translation
}
```

```python
for name, df in datasets.items():
    print(f"{name}")
    print(f"Shape : {df.shape}")
    print("-" * 40)
```

## Dataset Profiling

In this Section, we inspect each dataset to understand:

- Shape (rows)
- Column
- Data types
- Missing values
- Sample records

This helps identify potential data quality issues before cleaning.

```python
def inspect_dataset(name, df):
    print("=" * 70)
    print(f"DATASET: {name}")
    print("=" * 70)

    print(f"\nShape: {df.shape}")

    print(f"\nColumns:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nFirst Five Rows:")
    display(df.head())

    print("\n")
```

```python
for name, df in datasets.items():
    inspect_dataset(name, df)
```

```python
for name, df in datasets.items():
    print("=" * 70)
    print(name)
    print(df.describe(include="all"))
```

```python
for name, df in datasets.items():
    print(f"{name}: {df.memory_usage(deep=True).sum()/1024:.2f} KB")
```
